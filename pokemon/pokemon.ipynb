```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#read dataset
df = pd.read_csv('pokemon_data.csv', index_col= 0)
```

```python
df.head()
```

```python
df.describe(include='all')
```

```python
df.info()
```

```python
df['Type 2'].fillna('Unknown', inplace=True)
```

## Attack vs Defense

```python
# recommended way
sns.lmplot(x= 'Attack', y= 'Defense', data= df, fit_reg= False, hue= 'Generation')

# alternative way

#sns.lmplot(x= df.Attack, y= df.Defense)

# tweaking with matplotlib
plt.xlim(0, None)
plt.ylim(0, None)
```

# Interpretation

## Attributes plots

```python
# Boxplot
sns.barplot(data = df)
```

# Interpretation

## Attack by Type 1

```python
# Overlaying plots
plt.figure(figsize=(12,6)) 
sns.violinplot(x='Type 1',
               y='Attack', 
               data=df, 
               inner=None, # Remove the bars inside the violins
               palette=colors)
 

sns.swarmplot(x='Type 1', 
              y='Attack', 
              data=df, 
              color='k', # Make points black
              alpha=0.7) # and slightly transparent
 
# Set title with matplotlib
plt.title('Attack by Type 1')
```

# Interpretation

```python
melted_df= pd.melt(df,
                   id_vars=['Name', 'Type 1', 'Type 2', 'Generation', 'Legendary'],
                  var_name='Xp')
melted_df.head()
```

```python
# Pokemon color palette

colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

## Effect of Type 1 on Xp

```python
plt.figure(figsize=(20,6)) 

sns.barplot(x='Xp',               
              y='value',               
              data=melted_df,               
              hue='Type 1', palette= colors)  

plt.title('Effect of Type 1 on Xp')
plt.legend(bbox_to_anchor=(1, 1), loc=2)
```

# Interpretation

## Effect of Type 2 on Xp

```python
plt.figure(figsize=(20,6)) 

sns.barplot(x='Xp',               
              y='value',               
              data=melted_df,               
              hue='Type 2', palette=colors)  

plt.title('Effect of Type 2 on Xp')
plt.legend(bbox_to_anchor=(1, 1), loc=2)
```
